```python
#Write Code for Logistic Regression here

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LogisticRegression as LR
from sklearn.model_selection import train_test_split as tts
from sklearn.preprocessing import StandardScaler as SS

cancer_dataset = pd.read_csv('/content/dataR2.csv')

x = cancer_dataset[['Age','BMI','Glucose','Insulin','HOMA','Leptin','Adiponectin','Resistin','MCP.1']].values
y = cancer_dataset['Classification'].values

x_train,x_test,y_train,y_test = tts(x,y,test_size = 0.2)

scaler = SS()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

model = LR()

model.fit(x_train_scaled, y_train)

print(f"Model accuracy: {model.score(x_test_scaled, y_test) * 100}%")



```

```python
#Code for Linear Regression here

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression as LR
from sklearn.model_selection import train_test_split as tts

real_estate_dataset = pd.read_excel('/content/Real estate valuation data set.xlsx')

x = real_estate_dataset[['X1 transaction date', 'X2 house age','X3 distance to the nearest MRT station','X4 number of convenience stores','X5 latitude','X6 longitude']].values
y = real_estate_dataset['Y house price of unit area'].values

x_train,x_test,y_train,y_test = tts(x,y,test_size=0.2)

scaler = SS()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

model = LR()
model.fit(x_train_scaled,y_train)

pred = model.predict(x_test_scaled)

plt.scatter(y_test,pred, alpha=0.5)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', lw=2)
plt.xlabel('Actual House Prices')
plt.ylabel('Predicted House Prices')
plt.title('Actual vs. Predicted House Prices')
plt.show()

print(f"Model accuracy: {model.score(x_test_scaled, y_test) * 100}%")
```
